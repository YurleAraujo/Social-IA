**Transformación de los datos**

```python
#Se convierte la columna "date_posted" de object a datatime
```

```python
df_social['date_posted'] = pd.to_datetime(df_social['date_posted'], errors='coerce')
print('El tipo de columna de fecha de post: ', df_social['date_posted'].dtype)
```

```python
#se verifica que las columnas no tengas valores faltantes y reemplazarlos
```

```python
print (df_social['likes'].isna().sum())
df_social['likes'].fillna(0, inplace=True)
print(df_social['likes'].isna().sum())
```

```python
print (df_social['shares'].isna().sum())
df_social['shares'].fillna(0, inplace=True)
print(df_social['shares'].isna().sum())
```

```python
print (df_social['comments'].isna().sum())
df_social['comments'].fillna(0, inplace=True)
print(df_social['comments'].isna().sum())
```

```python
print (df_social['date_posted'].isna().sum())
df_social['date_posted'].fillna(0, inplace=True)
print(df_social['date_posted'].isna().sum())
```

```python
print (df_social['sentiment_score'].isna().sum())
df_social['sentiment_score'].fillna(0, inplace=True)
print(df_social['sentiment_score'].isna().sum())
```

```python
#Se obtiene nuevamente la información estadistica y resumida del df_social limpio
```

```python
df_social.describe()
```

**Graficas**

```python
for i_grafica in ['platform', 'ai_topic', 'location']:
    sns.histplot(data=df_social, x=i_grafica, kde=True)
    plt.ylabel('Cantidad de publicaciones')
    plt.xticks(rotation=45)
   
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation = df_social[['likes', 'shares', 'comments', 'sentiment_score']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación entre variables')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_social, x='ai_topic', hue='location', palette='Set2')

plt.title('Temas de IA vs Países')
plt.xlabel('Tópicos de IA')
plt.ylabel('Cantidad de publicaciones')
plt.xticks(rotation=45)
plt.legend(title='País')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_social, x='platform', hue='location', palette='Set2')

plt.title('Plataforma vs Países')
plt.xlabel('Tópicos de IA')
plt.ylabel('Cantidad de publicaciones')
plt.xticks(rotation=45)
plt.legend(title='País')
plt.tight_layout()
plt.show()
```
